```python
import numpy as np
import pandas as pd
df1=pd.read_excel('Ticket Dump 14-Nov-20 V1.0.xlsx')
X1=df1['Short description']

X2=""
for i in range(0,len(X1)):
    X2=X2+str(" ") + str(X1[i])
    
X3=X2.lower()
```

```python
from nltk.tokenize import word_tokenize

token1=word_tokenize(X3)
```

```python
from nltk.util import bigrams,trigrams,ngrams
from nltk.probability import FreqDist
```

```python
quotes_bigram1=list(nltk.bigrams(token1))
```

```python
quotes_trigram1=list(nltk.trigrams(token1))
```

```python
quotes_ngrams1=list(nltk.ngrams(token1,10))
```

```python
Fdist11=FreqDist()

Fdist_token11=quotes_bigram1

for word in Fdist_token11:
    Fdist11[word]+=1
Fdist11
```

```python
Fdist12=FreqDist()

Fdist_token12=quotes_trigram1

for word in Fdist_token12:
    Fdist12[word]+=1
Fdist12
```

```python
Fdist13=FreqDist()

Fdist_token13=quotes_ngrams1

for word in Fdist_token13:
    Fdist13[word]+=1
Fdist13
```

```python
## CHange if Punctuation and Stopwords are removed

```

```python
from nltk.corpus import stopwords
stopwords1=set(stopwords.words('english'))
```

```python
import re
#punctuation=re.compile(r'[-.?!,:;()|0-9]')
punctuation=re.compile(r'[-.?!,+&:;()@#|/]')
```

```python
post_punctuation=[]
for words in token1:
    word=punctuation.sub("",words)
    if len(word)>0:
        post_punctuation.append(word)
```

```python
token2=post_punctuation
```

```python
filter2=[]

for w in token2:
    if w not in stopwords1:
        filter2.append(w)

print(filter2)
```

```python
quotes_ngrams2=list(nltk.ngrams(filter2,4))
quotes_ngrams2
```

```python
Fdist21=FreqDist()

Fdist_token21=quotes_ngrams2

for word in Fdist_token21:
    Fdist21[word]+=1
Fdist21
```
